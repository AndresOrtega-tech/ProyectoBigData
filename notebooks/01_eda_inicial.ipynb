## Objetivos
1. Confirmar las dimensiones del dataset (500 filas esperadas)
2. Verificar valores faltantes/nulos
3. Calcular rango de cada variable (min, max, promedio)

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.precision', 2)

print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
df = pd.read_csv('../data/ai_dev_productivity.csv')
print("Dataset cargado correctamente.")
```

---
## 1. ¿Cuántos datos tenemos?

```python
filas, columnas = df.shape
print(f"Filas: {filas}")
print(f"Columnas: {columnas}")

print(f"\nColumnas del dataset:")
for i, col in enumerate(df.columns, 1):
    print(f"  {i}. {col}")
```

```python
df.head(10)
```

---
## 2. ¿Hay valores faltantes/nulos?

```python
print("Valores nulos por columna:")
nulos = df.isnull().sum()
print(nulos)

total_nulos = nulos.sum()
print(f"\nTotal de valores nulos en el dataset: {total_nulos}")

if total_nulos == 0:
    print("\nEl dataset esta completo, no hay valores faltantes.")
else:
    print(f"\nSe encontraron {total_nulos} valores faltantes.")
    print("\nPorcentaje de nulos por columna:")
    print((nulos / len(df) * 100).round(2))
```

---
## 3. ¿Qué rango tiene cada variable? (min, max, promedio)

```python
df.describe()
```

```python
resumen = pd.DataFrame({
    'Min': df.min(),
    'Max': df.max(),
    'Promedio': df.mean(),
    'Mediana': df.median(),
    'Rango': df.max() - df.min()
})

print("Resumen de rangos por variable:")
resumen
```
